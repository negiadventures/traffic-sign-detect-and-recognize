## NOTE before start training: 
1. --data should be passed abosolute path to the signs.yaml file which is typically inside dataset directory
2. --cfg should be passed abosolute path to the models/yolov5.yaml which is inside the yolov5 repository into which we entered in previous step
3. --weights should be passed yolov5s.pt file if running the model for the first time which is in the yolov5 repository and later if you want to continue the model from previous state, you can pass best.pt which is inside --project location in weights directory.
4. --project is the directory created for your model inside yolov5 repository

```python
# Train YOLOv5s for 100 epochs
!python train.py --img 640 --batch 16 --epochs 100 --data /common/home/an721/Downloads/yolo_sign/dataset/signs.yaml --cfg /common/home/an721/Downloads/yolo_sign/yolov5/models/yolov5s.yaml --weights newsigns/signs5/weights/best.pt --name signs --project newsigns
```
